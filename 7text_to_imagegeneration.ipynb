```python
# Import necessary libraries
from diffusers import DiffusionPipeline  # Stable Diffusion pipeline for text-to-image generation
import torch  # For tensor computations
import gradio as gr  # Gradio for building the UI
```

```python
# Initialize the Stable Diffusion pipeline
pipe = DiffusionPipeline.from_pretrained(
    "stabilityai/stable-diffusion-xl-base-1.0",
    torch_dtype=torch.float16,  # Use half precision for better performance
    use_safetensors=True,  # Ensure safe tensor computations
    variant="fp16"  # Float16 variant for optimized computation
)
pipe.to("cuda")  # Move the model to GPU (CUDA) for faster computation
```

```python
# Define the function to generate an image from a prompt
def generate_image(prompt, negative_prompt):
    # Generate image using the diffusion pipeline
    image = pipe(prompt=prompt, negative_prompt=negative_prompt).images[0]
    return image  # Return the generated image
```

```python
# Define a function to reset the input fields
def restart_inputs():
    return "", ""  # Clear the prompt and negative prompt
```

```python
# Custom CSS styling for Gradio interface
custom_css = """
#prompt-input {
    font-family: 'Arial', sans-serif;
    font-size: 18px;
    padding: 10px;
}
#output-image {
    border: 2px solid #ccc;
    border-radius: 10px;
    margin-top: 10px;
}
.gr-button {
    background-color: #007bff;
    color: white;
    border: none;
    padding: 10px 20px;
    font-size: 16px;
    cursor: pointer;
}
.gr-button:hover {
    background-color: #0056b3;
}
"""
```
